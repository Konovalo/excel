```python
import pandas as pd
data = pd.read_csv("telecom_churn.csv")
data.head()
```

```python
# разделим таблицу на x - целевая переменная и y - признаки
x = data.drop(columns = "Churn")
y = data["Churn"]
```

```python
# разделим значения на тренирововчные, тестовые и валидационные данные.
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size = 0.25)
# Делаем деление на тренировочную и валидационную часть, а тестовыю не делаем (в этой задаче это не надо)
```

```python
from sklearn import neighbors
# Создадим алгоритм машинного обучения k-ближайших соседей
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
# указываем в скобках гипер параметры (указываем, что соседей котрых мы учитываем 5)
knn
```

```python
# метод knn.fit(X_train, y_train) принимает два аргумента
# x_train матрица в каждой клетке которой лежит значение признака объекта
# y_train вектор в каждой клетке которого лежит значение 1 или 0, клиент остался или ушел
knn.fit(x_train, y_train)
```

```python
# Предсказание результатов: preds_valid = knn.predict(X_valid)
# внутрь в качестве одного аргцмента передаем признаки объектов для которых надо сделать предсказание
preds_valid = knn.predict(x_valid)
preds_valid
```

```python
# Посмотрим на сколько точная получилась наша модель, посчитаем accuracy
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score
print(f'Accuracy={accuracy_score(y_valid, preds_valid)}')
# несмотря на хороший результат оказывается, что мы редложилибы скидку почти в два раза большим людям,
# чем надо было бы, а действиетльно ушедших было только 27% с кем мы поработали
```
